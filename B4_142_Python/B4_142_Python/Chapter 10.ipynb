```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(10,10))
x=np.array([28,27,6,6])
mylabels=["Apple","Mango","Banana","Pear"]
myexplode=[0,0.2,0,0]
colors=["red","orange","yellow","green"]
plt.pie(x,startangle=90,labels=mylabels,explode=myexplode,shadow=True,colors=colors,autopct="%1.2f%%")
plt.title("Pie Chart")
plt.legend(loc="best",title="Fruits")
plt.tight_layout()
plt.show()
```

```python
#Create a pie chart for popularity data of different programming language and display it as a pie chart using matplotlib
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(10,10))
python=29
java=19
csharp=7
php=6
cpp=5
r=3
x=np.array([python,java,csharp,php,cpp,r])
mylabels=np.array(["python","java","csharp","php","cpp","r"])
myexplode=[0.3,0,0,0,0,0]
plt.pie(x,labels=mylabels,explode=myexplode,autopct="%1.2f%%",shadow=True)
plt.legend(loc="best",title="Programming Language")
plt.show()
```
